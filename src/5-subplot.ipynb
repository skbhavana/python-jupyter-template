### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Line plot (First subplot)
x = np.linspace(0, 10, 100)
y = x ** 2
axes[0, 0].plot(x, y, color='b', label='y = x^2')
axes[0, 0].set_title('Line Chart')
axes[0, 0].set_xlabel('X values')
axes[0, 0].set_ylabel('Y values')
axes[0, 0].grid(True)
axes[0, 0].legend()

# Bar chart (Second subplot)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axes[0, 1].set_title('Bar Chart')
axes[0, 1].set_xlabel('Categories')
axes[0, 1].set_ylabel('Values')
axes[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

# Scatter plot (Third subplot)
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axes[1, 0].scatter(x_scatter, y_scatter, color='orange')
for i in range(len(x_scatter)):
    axes[1, 0].annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='center')
axes[1, 0].set_title('Scatter Plot')
axes[1, 0].set_xlabel('X values')
axes[1, 0].set_ylabel('Y values')
axes[1, 0].grid(True)

# Pie chart (Fourth subplot)
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['blue', 'red', 'green', 'orange']
axes[1, 1].pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
axes[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
