### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
y = x ** 2  # Square of x

# Create the line plot
plt.plot(x, y, label='y = x^2', color='b')

# Customizing the line plot
plt.title('Line Plot of y = x^2')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True)
plt.legend()

# Show the line plot
plt.show()

# Data for bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

# Create the bar plot
plt.bar(categories, values, color=colors)

# Customizing the bar plot
plt.title('Bar Plot of Categories')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the bar plot
plt.show()
```
